## Tokenization in NLP

**1. Corpus:** -  A paragraph given
**2. Documents** - A Sentences
**3. Vocabulary** - A Unique Words
**4. Words** -  All the possible unique words in the corpus


**Tokenization**: 
Paragarph -> Sentences 
OR
Sentences -> Words




# Tokenization Practicals

```python

corpus = """ Hello Welcome, to Krish Naik's NLP Tutorials.
Please do watch the entire course! to become the expert in NLP.
"""
```

```python
print(corpus)
```

```python
## Tokenization
## Sentence to paragraph

from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
## Tokenization 
# Paragarph --> Words
# Sentence --> Words

from nltk.tokenize import word_tokenize
words = word_tokenize(corpus)
```

```python
words
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```

# Text Pre-processing Stemming Using NLTK


**Stemming**: Process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as lemma.


```python
words = [
    # Basic verbs and plurals
    "connect", "connected", "connecting", "connections", "connects",
    "fences", "dictated", "beholden",
    
    # The 'Y' rule
    "happy", "happily", "happiness", "sky", "skies", "duty", "duties",
    
    # Double consonants and vowel rules
    "hop", "hopping", "hopped", "hope", "hoping", "hoped",
    "tan", "tanned", "tanning",
    
    # Derivational suffixes (-ation, -al, -ness, -ment)
    "national", "nationalization", "nationwide",
    "adjustment", "adjusting", "adjustable",
    "relational", "relationship", "relate",
    
    # Complex or irregular cases (to see how your stemmer handles them)
    "university", "universe", "universal",
    "organization", "organize", "organ",
    "knives", "feet", "mice"
]
```

```python
## PORTER STEMMER

from nltk.stem import PorterStemmer

stemming = PorterStemmer()

for word in words:
    print(f"{word}====={stemming.stem(word)}")
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')
```

## RegexpStemmer Class

NLTK has RegexpStemmer class with the help of which we can easily implemnt regular expressions. It basically take a single regular expression and removes any prefix or suffix that matches the expression. 

```python
from nltk.stem import RegexpStemmer
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$',min=4)

reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')
```

## Snow Ball Stemmer
Better in Accurarcy than the Potter Stemmer


```python
from nltk.stem import SnowballStemmer

snow_stem = SnowballStemmer('english')
```

```python
for word in words:
    print(f"{word} ====== {snow_stem.stem(word)}")
```

```python
snow_stem.stem('goes')
```
